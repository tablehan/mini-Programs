```python
import requests
from bs4 import BeautifulSoup
import json
import pandas as pd
import numpy as np
import time
import csv
import gc
```

```python
url="https://grbdef.stpi.narl.org.tw/searcher"
headers = {
    "User-Agent":'Mozilla/5.0 (Windows NT 6.1; Win64; x64; rv:57.0) Gecko/20100101 Firefox/57.0',
#     "Accept":"application/json, text/plain, */*",
#     "Accept-Encoding":"gzip, deflate, br",
#     "Accept-Language":"zh-TW,zh;q=0.8,en-US;q=0.5,en;q=0.3",
#     "Cache-Control":"no-cache",
#     "Connection":"keep-alive",
#     "Content-Length":"590",
#     "Content-Type":"application/x-www-form-urlencoded;charset=UTF-8",
#     "Host":"grbdef.stpi.narl.org.tw",
#     "Origin":"https://www.grb.gov.tw",
#     "Pragma":"no-cache",
    "Referer":"https://www.grb.gov.tw/", # 這個一定要放
#     "Sec-Fetch-Dest":"empty",
#     "Sec-Fetch-Mode":"no-cors",
#     "Sec-Fetch-Site":"cross-site",
}

data={
    "keyword": "",
    "b": "",
    "queryType": "GRB05",
    "projNums": "",
    "planYearSt": "82",
    "planYearEn": "111",
    "actYearSt": "82",
    "actYearEn": "111",
#     "planAmtSt": "",
#     "planAmtEn": "",
    "memberKeywordScope": "0",
    "organKeywordScope": "0",

    "nowPage": "1",
    "rowsPerPage": "200",
    "orderType": "SIMILARITY",
    "codeWiat1": "true",
    "codeWiat2": "true",
    "codeWiat3": "true",
    "techName": "true",
    "techDesc": "true",
    "baseSeq": "",
    "organCode": "",
    "trl1": "false",
    "trl2": "false",
    "trl3": "false",
    "trl4": "false",
    "trl5": "false",
    "trl6": "false",
    "trl7": "false",
    "trl8": "false",
    "trl9": "false",
    "trl0": "false"
}
def get_by_page(i):
    data["nowPage"]=str(i)
    return data
    
```

```python
response = requests.post(url, headers = headers, data = get_by_page(1))
```

```python
x=json.loads(response.text)
totalRows=x['totalRows']
totalPages=x['totalPages']
start=1
```

```python
res=[]
for i in range (start,totalPages+1):
    try:
        page_res=requests.post(url, headers = headers, data = get_by_page(i))
        temp_res=json.loads(page_res.text)
        if len(temp_res['obj'])>0:
            for obj in temp_res['obj']:
                res.append([i,obj['id'],obj['title'],len(obj['hostNameList']),obj['planAmt']])
            print(f'page{i}: {time.strftime("%Y-%m-%d %I:%M:%S %p", time.localtime())}')
            time.sleep(0.2)
        else:
            print(f"No data: finished page:{i}")
            break
    except:
        print(f'Error, finished page:{i}')
print('crewler finished')
```

```python
with open('outputGRBid2.csv', 'w', newline='', encoding='UTF-8') as csvfile:
  # 建立 CSV 檔寫入器
    writer = csv.writer(csvfile)
    writer.writerow(["page","id","title","numPro","planAmt"])
    for r in res:
        writer.writerow(r)
```
